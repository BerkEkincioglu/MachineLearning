```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_csv('random-forest-regression-dataset.csv',sep=';',header=None)
```

```python
df.head()
```

```python
x = df.iloc[:,0].values.reshape(-1,1)
y = df.iloc[:,1].values.reshape(-1,1)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators=100,random_state=42)
rf_reg.fit(x,y)
plt.scatter(x,y,color='orange')
plt.plot(x,rf_reg.predict(x),color='black')
plt.show()
```

```python
x_ = np.arange(min(x),max(x),0.01).reshape(-1,1)
```

```python
plt.scatter(x,y,color='orange')
plt.plot(x_,rf_reg.predict(x_),color='blue')
plt.show()
```

```python
plt.scatter(x,y,color='orange')
plt.plot(x_,rf_reg.predict(x_),color='blue')
plt.plot(x,rf_reg.predict(x),color='black')
plt.show()
```

```python
print("7.7 seviyesinde fiyatın ne kadar olduğu: ",rf_reg.predict([[7.7]]))
```
