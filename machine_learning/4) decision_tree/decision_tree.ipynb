```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('/Users/berkekincioglu/Desktop/machine_learning/decision_tree/decision-tree-regression-dataset.csv',sep=';',header=None)
```

```python
df
```

```python
x = df.iloc[:,0:1].values.reshape(-1,1)
y = df.iloc[:,1:].values.reshape(-1,1)
```

```python
from sklearn.tree import DecisionTreeRegressor
r_df = DecisionTreeRegressor()
r_df.fit(x,y)
x_ = np.arange(min(x),max(x),0.01).reshape(-1,1)
y_head = r_df.predict(x_)
```

```python
plt.scatter(x,y,color='orange')
plt.plot(x,r_df.predict(x),color='black')
plt.xlabel('tribun level')
plt.ylabel('price')
plt.show()
```

```python
plt.scatter(x,y,color='orange')
plt.plot(x_,y_head,color='black')
plt.xlabel('tribun level')
plt.ylabel('price')
plt.show()
```
