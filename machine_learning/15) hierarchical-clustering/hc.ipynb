```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# class1
x1 = np.random.normal(25,5,100)
y1 = np.random.normal(25,5,100)
# class2
x2 = np.random.normal(55,5,100)
y2 = np.random.normal(60,5,100)
# class3
x3 = np.random.normal(55,5,100)
y3 = np.random.normal(15,5,100)

x = np.concatenate((x1,x2,x3),axis=0)
y = np.concatenate((y1,y2,y3),axis=0)

d = {'x':x,'y':y}

df = pd.DataFrame(d)

plt.scatter(x1,y1,c='black')
plt.scatter(x2,y2,c='black')
plt.scatter(x3,y3,c='black')
plt.show()
```

```python
# Dendrogram

from scipy.cluster.hierarchy import linkage,dendrogram

merg = linkage(df,method='ward')
dendrogram(merg,leaf_rotation=90)
plt.xlabel('data points')
plt.ylabel('euclidean distance')
plt.show()
```

```python
# HC

from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
cluster = ac.fit_predict(df)
df['label'] = cluster 
plt.scatter(df.x[df.label == 0],df.y[df.label == 0],c='orange')
plt.scatter(df.x[df.label == 1],df.y[df.label == 1],c='black')
plt.scatter(df.x[df.label == 2],df.y[df.label == 2],c='grey')
plt.show()
```
