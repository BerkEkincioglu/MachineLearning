```python
from sklearn.datasets import load_iris
import pandas as pd 
import numpy as np 
```

```python
iris = load_iris()
data = iris.data
feature_names = iris.feature_names
y = iris.target
df = pd.DataFrame(data=data,columns=feature_names)
df['class'] = y
x = data
```

```python
# PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2,whiten=True) #whiten --> normalization, n_components= dimension figure
pca.fit(x)

x_pca = pca.transform(x)

print('variance ratio : ',pca.explained_variance_ratio_)
print('sum: ', sum(pca.explained_variance_ratio_))

# we lost 3% of data
```

```python
# visualization 2D

df['p1'] = x_pca[:,0]
df['p2'] = x_pca[:,1]

colors = ['black','orange','grey']
import matplotlib.pyplot as plt 
for i in range(3):
    plt.scatter(df.p1[df['class'] == i],df.p2[df['class'] == i], color=colors[i], label = iris.target_names[i])
```
