```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df_lin = pd.read_csv('linear-regression-dataset.csv',sep=';')
df_rf = pd.read_csv('random-forest-regression-dataset.csv',sep=';')
```

```python
# R^2 with linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# visualization
plt.scatter(df_lin.deneyim,df_lin.maas,color='orange')
plt.xlabel('deneyim')
plt.ylabel('maas')
plt.show()
```

```python
x = df_lin.maas.values.reshape(-1,1)
y = df_lin.deneyim.values.reshape(-1,1)
```

```python
lr.fit(x,y)
plt.scatter(x,y,color='orange')
plt.plot(x,lr.predict(x),color='black')
```

```python
y_head = lr.predict(x)
from sklearn.metrics import r2_score
r2 = r2_score(y,y_head)
r2
```

```python
x2 = df_rf.iloc[:,0].values.reshape(-1,1)
y2 = df_rf.iloc[:,1].values.reshape(-1,1)
```

```python
# R2 with random forest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x2,y2)
y_head = rf.predict(x2)
```

```python
plt.scatter(x2,y2,color='orange')
plt.plot(x2,y_head,color='black')
```

```python
r2 = r2_score(y2,y_head)
r2
```
